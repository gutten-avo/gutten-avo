#  Apa itu Rasio Peserta Didik dan Fasilitas?

* Rasio ini menghitung berapa banyak siswa yang harus berbagi satu fasilitas.

#  Tujuannya:
* Untuk menormalkan data, karena jumlah siswa di tiap daerah berbeda.
* Agar analisis lebih adil: misalnya, 1 laboratorium untuk 1.000 siswa di daerah A  1 lab untuk 100 siswa di daerah B.

**Nah, rasio itu artinya:**
Kita hitung berapa banyak murid yang harus berbagi satu fasilitas.

* Kalau angkanya kecil, itu bagus  artinya muridnya dapat bagian fasilitas yang lebih banyak.
* Kalau angkanya besar, itu kurang bagus  berarti fasilitasnya kurang untuk jumlah muridnya.

```python
import numpy as np

# Salin dataframe untuk menjaga data asli
df_rasio = df_merged_final.copy()

# Rasio Lab IPA
df_rasio['rasio_labor'] = np.where(
    df_rasio['Jumlah Lab IPA'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik'] / df_rasio['Jumlah Lab IPA']
)

# Rasio Perpustakaan
df_rasio['rasio_perpus'] = np.where(
    df_rasio['Jumlah Perpustakaan'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik'] / df_rasio['Jumlah Perpustakaan']
)

# Rasio Toilet Laki-laki
df_rasio['rasio_toilet_l'] = np.where(
    df_rasio['Jumlah Toilet Laki-laki'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik (L)'] / df_rasio['Jumlah Toilet Laki-laki']
)

# Rasio Toilet Perempuan
df_rasio['rasio_toilet_p'] = np.where(
    df_rasio['Jumlah Toilet Perempuan'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik (P)'] / df_rasio['Jumlah Toilet Perempuan']
)

# Rasio UKS
df_rasio['rasio_uks'] = np.where(
    df_rasio['Jumlah UKS'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik'] / df_rasio['Jumlah UKS']
)

# Rasio Ruang Kelas
df_rasio['rasio_kelas'] = np.where(
    df_rasio['Jumlah Ruang Kelas'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik'] / df_rasio['Jumlah Ruang Kelas']
)

# Rasio Tenaga Pengajar
df_rasio['rasio_pengajar'] = np.where(
    df_rasio['Jumlah Tenaga Pengajar'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik'] / df_rasio['Jumlah Tenaga Pengajar']
)

# Rasio Sumber Air
df_rasio['rasio_air'] = np.where(
    df_rasio['Jumlah Sumber Air'] == 0,
    0,
    df_rasio['Jumlah Peserta Didik'] / df_rasio['Jumlah Sumber Air']
)

# Rasio Putus Sekolah
df_rasio['rasio_putus'] = np.where(
    df_rasio['Jumlah Peserta Didik'] == 0,
    0,
    df_rasio['Jumlah Putus Sekolah'] / df_rasio['Jumlah Peserta Didik']
)
```

# Elbow Curve Method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Menentukan jumlah cluster optimal dengan Elbow Method
wcss = []  # Within-Cluster Sum of Squares

for i in range(1, 11):  # Uji dari 1 sampai 10 cluster
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=100, n_init=10, random_state=0)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Visualisasi Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method - Menentukan Jumlah Cluster')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

# K Means

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Ganti df_clean dengan df_scaled kamu
df_clean = df_scaled.copy()

# Tentukan jumlah cluster optimal (hasil elbow curve kamu: 4)
optimal_clusters = 4

# Inisialisasi dan fit KMeans
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(df_clean)

# Visualisasi clustering tanpa PCA
plt.figure(figsize=(10, 6))
plt.scatter(df_clean.values[y_kmeans == 0, 0], df_clean.values[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(df_clean.values[y_kmeans == 1, 0], df_clean.values[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(df_clean.values[y_kmeans == 2, 0], df_clean.values[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(df_clean.values[y_kmeans == 3, 0], df_clean.values[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')

# Titik centroid
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='yellow', label='Centroids')

plt.title('Clusters of Data (Before PCA)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# Hitung skor Silhouette
score = silhouette_score(df_clean, y_kmeans)
print(f'Silhouette Score: {score:.4f}')
```

# PCA (Reduction Dimensional)
PCA digunakan untuk mengecilkan dimensi agar hasil clustering lebih akurat dan efisien.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# PCA awal untuk mengetahui jumlah komponen dengan cumulative explained variance 80%
pca = PCA()
df_pca = pca.fit_transform(df_clean)

# Hitung cumulative explained variance (CEV)
cev = pca.explained_variance_ratio_.cumsum()
n_components = np.argmax(cev >= 0.8) + 1  # Ambil komponen minimum untuk 80% informasi

print("Jumlah komponen setelah mengambil minimal 80% cumulative variance:", n_components)
print("Cumulative explained variance:", cev)

# Gunakan jumlah komponen optimal untuk reduksi dimensi
pca_final = PCA(n_components=n_components)
df_reduced = pca_final.fit_transform(df_clean)

# Uji kembali Elbow Curve setelah PCA
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=100, n_init=10, random_state=0)
    kmeans.fit(df_reduced)
    wcss.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 10), wcss, marker='o', linestyle='--')
plt.title('Elbow Curve (Setelah PCA)')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Gunakan jumlah cluster optimal (misalnya 4)
optimal_clusters = 4

# Lakukan clustering
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(df_reduced)

# Visualisasi hasil clustering (pakai 2 dimensi pertama dari PCA)
plt.figure(figsize=(10, 6))
plt.scatter(df_reduced[y_kmeans == 0, 0], df_reduced[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(df_reduced[y_kmeans == 1, 0], df_reduced[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(df_reduced[y_kmeans == 2, 0], df_reduced[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(df_reduced[y_kmeans == 3, 0], df_reduced[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')

# Centroid-nya
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Data (Setelah PCA)')
plt.xlabel('PCA Komponen 1')
plt.ylabel('PCA Komponen 2')
plt.legend()
plt.grid(True)
plt.show()

# Skor evaluasi
score = silhouette_score(df_reduced, y_kmeans)
print(f'Silhouette Score (Setelah PCA): {score:.4f}')
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Inisialisasi dan jalankan DBSCAN
dbscan = DBSCAN(eps=1, min_samples=10)
cluster_dbscan = dbscan.fit_predict(df_reduced)

# Hitung jumlah cluster yang terbentuk
n_clusters = len(set(cluster_dbscan)) - (1 if -1 in cluster_dbscan else 0)
print(f"Jumlah cluster (tidak termasuk noise): {n_clusters}")

# Hitung Silhouette Score jika jumlah cluster > 1
if n_clusters > 1:
    score = silhouette_score(df_reduced, cluster_dbscan)
    print(f"Silhouette Score: {score:.3f}")
else:
    print("Tidak dapat menghitung Silhouette Score karena hanya ada satu cluster atau semuanya noise.")

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(df_reduced[:, 0], df_reduced[:, 1], c=cluster_dbscan, cmap='plasma', s=50)
plt.title("Hasil Clustering dengan DBSCAN")
plt.xlabel("PCA Komponen 1")
plt.ylabel("PCA Komponen 2")
plt.colorbar(label='Cluster')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Buat linkage matrix
linked = linkage(df_reduced, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.title("Dendrogram untuk Hierarchical Clustering")
plt.xlabel("Indeks Data")
plt.ylabel("Jarak")
plt.show()

# Lakukan clustering
model = AgglomerativeClustering(n_clusters=4, linkage='ward')
labels = model.fit_predict(df_reduced)

# Visualisasi hasil clustering
plt.figure(figsize=(8, 6))
plt.scatter(df_reduced[:, 0], df_reduced[:, 1], c=labels, cmap='viridis')
plt.title("Hasil Hierarchical Clustering")
plt.xlabel("PCA Komponen 1")
plt.ylabel("PCA Komponen 2")
plt.colorbar(label="Cluster")
plt.show()

# Evaluasi clustering
score = silhouette_score(df_reduced, labels)
print(f"Silhouette Score: {score:.3f}")
print(f"Jumlah Cluster: {len(np.unique(labels))}")
```
